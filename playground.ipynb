Order lifecycle Simulator

```python
from order import Order
from order import OrderState


symbol = 'AAPL'
qty = 100
side = 'BUY'

test_order = Order(symbol, qty, side)

print(test_order.state)

test_order.transition(OrderState.ACKED)
print(test_order.state)

try:
    test_order.transition(OrderState.NEW)
except ValueError as e:
    print(e)

print(test_order.state)
```
